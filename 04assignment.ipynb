### How do we categorize a point? Measure a point? We look into the coordinate system in various formalisms
We categorize a point with the help of dimensions which coressponds the number of coordinates needed to specify it's location. A point in 1-Dimension cartesian coordinate system has one coordinate (x). A point in 2-Dimension has coordinates (x, y) while a point in 3-Dimension has coordinates (x, y, z) and in a Hyper-Dimension coordinates will be (x1, x2, x3,...., xn).
We can measure a points distance from origin using in 1-D using formula in coordinate (x)
as |x|= √x². In 2-Dimension having coordinates (x, y) we can measure it using formula |x|=√(x²+y²). In 3-Dimension having coordinates (x, y, z) we can measure it using formula |x|=√(x²+y²+z²). In Hyper-Dimension having coordinates (x, x1, x2, x3,...xn) we can measure it using formula |x|=√(x1²+x2²+x3²+....+xn²).

##### The angle between the vectors joining the origin with those points

```python
import math as m

# Function finds the distance between two points
def angle_b_two_points(*p):
    x1,x2=p[0] #Coordinate of x1 and x2
    y1,y2=p[1] #Coordinate of y1 and y2

    # Calculates product sum
    product_sum=x1*y1+x2*y2

    # Calculates magitude
    magitude_x=m.sqrt((x1)**2+(x2)**2)
    magitude_y=m.sqrt((y1)**2+(y2)**2)
    magnitude=magitude_x*magitude_y
    
    # Calculate angle
    angle=m.acos(product_sum/magnitude)

    # Convert radians into degrees and rounding 
    degrees=round(angle*((180*7)/22),1)
    return degrees
# Empty list to store userinput coordinates for x and y point
pointx=[]
pointy=[]

# Empty list to store pointx and pointy
point=[]

# User input size of dimesion
dimension=int(input('Enter the dimensions:'))

# User input coordinates for x point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate of x{coordinates+1}:'))
    pointx.append(new_point) # Appends every new coordinate value entered by User
point.append(pointx)

# User input coordinates for y point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate of y{coordinates+1}:'))
    pointy.append(new_point) # Appends every new coordinate value entered by User
point.append(pointy)

print(f'Coordinates are:{point}')
# Function call and storing value in result
result=angle_b_two_points(*point)

# Show result
print(f'Angle between two points is: {result}°')
```

```python
# Function finds equation of the plane is perpendicular to the line joining the two points and bisecting it
def equation_of_plane(*p):
    
    x1,y1,z1=p[0] #Coordinate of x1 ,y1 and z1
    x2,y2,z2=p[1] #Coordinate of x2, y2 and z2
    
    # Calculate midpoints
    mid_pointx=(x1+x2)//2
    mid_pointy=(y1+y2)//2
    mid_pointz=(z1+z2)//2

    # Calculate directonal vectors
    vector_dx=x2-x1
    vector_dy=y2-y1
    vector_dz=z2-z1
    
    # Equation of plane
    return f'{vector_dx}(x-{mid_pointx})+{vector_dy}(y-{mid_pointy})+{vector_dz}(z-{mid_pointz})=0'

# Empty list to store userinput coordinates for x,y point
pointx=[]
pointy=[]

# Empty list to store pointx and pointy
point=[]
# User input size of dimesion
dimension=int(input('Enter the dimensions:'))

# User input coordinates for x point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate{coordinates+1} of point1:'))
    pointx.append(new_point) # Appends every new coordinate value entered by User
point.append(pointx)

# User input coordinates for y point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate{coordinates+1} of point2:'))
    pointy.append(new_point) # Appends every new coordinate value entered by User
point.append(pointy)

print(f'Coordinates are:{point}')
# Function call and storing value in result
result=equation_of_plane(*point)

# Show result
print(f'Equation of plane is :{result}')
```

```python
# Function calculates projection of one vector on the other
def projection(*p):
    x1,x2=p[0] #Coordinates of x1 and x2
    y1,y2=p[1] #Coordinates of y1 and y2

    # Calculates product sum or dot product
    product_sum=x1*y1+x2*y2
    
    # Calculates norm
    norm_y=(y1)**2+(y2)**2
    
    # Calculates scalar projection
    projection_of_x_onto_y=product_sum/norm_y

    # Calculates vector projection 
    p1=projection_of_x_onto_y*y1
    p2=projection_of_x_onto_y*y2
    return f'{p1}i,{p2}j'
    
# Empty list to store userinput coordinates for x,y point
pointx=[]
pointy=[]

# Empty list to store pointx and pointy
point=[]

# User input size of dimesion
dimension=int(input('Enter the dimensions:'))

# User input coordinates for x point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate of x{chr(coordinates+105)}:'))
    pointx.append(new_point) # Appends every new coordinate value entered by User
point.append(pointx)

# User input coordinates for y point
for coordinates in range(dimension):
    new_point=int(input(f'Enter Coordinate of y{chr(coordinates+105)}:'))
    pointy.append(new_point) # Appends every new coordinate value entered by User
point.append(pointy)

print(f'Coordinates are:{point}')
# Function call and storing value in result
result=projection(*point)

# Show result
print(f'Projection of {pointx} on {pointy} is:{result}')
```

##### Given an equation of line as input see whether the line is parallel to the plane or not

```python
import math as m
# Function checks line is parallel to the plane.
def line_paralle_to_plane(dir_vector,norm_vector):

    x1,x2,x3=dir_vector # Vector x element x1 and x2
    y1,y2,y3=norm_vector # Vector y element y1 and y2
    # Calculates product sum
    product_sum=x1*y1+x2*y2+x3*y3

    # Calculates magitude
    magitude_x=m.sqrt((x1)**2+(x2)**2+(x3)**2)
    magitude_y=m.sqrt((y1)**2+(y2)**2+(y3)**2)
    magnitude=magitude_x*magitude_y
    
    # Calculate angle between line and plane
    angle=m.acos(product_sum/magnitude)*(180*7/22)
    if(angle==0):
        return 'Parallel'
    else:
        return 'not Parallel'
        
# Empty list to store elements of vector
vector_d=[]
vector_n=[]

# User input size of dimesion
dimension=int(input('Enter the dimensions:'))    

# User input direction vectors for line
for element in range(dimension):
    new_element=int(input(f'Enter direction vector {chr(element+97)} of line:'))
    vector_d.append(new_element) # Appends every new vector element value entered by User

# User input direction vectors for plane
for element in range(dimension):
    new_element=int(input(f'Enter normal vector {chr(element+65)} of plane:'))
    vector_n.append(new_element) # Appends every new vector element value entered by User

# Function call and storing value in result
result=line_paralle_to_plane(vector_d,vector_n)

# Show result
print(f'Line is {result} to the plane.')
```

### b. The Polar and Cylindrical Coordinate System
#### b.1 Conversion of 2D coordinates
##### Given a point in cartesian coordinates write a program to convert it to polar coordinates an do the reverse.

```python
import math as m

# Functionc calculates polar coordinates from cartesian coordinates
def cartesian_to_polar(*c):
    x,y=c # cartesain coordinates

    # Calculates radian
    r=round(m.sqrt(x**2+y**2),2)

    # Calculates theta or angle
    theta=round((m.atan(y/x))*(180/m.pi),2)

    return r,theta

#  Function calculates cartesian coordinates from polar coordinates
def polar_to_cartesian(*r):
    r,theta=r[0],r[1] # polar coordinates
   
    # Calculates x and y coordinate
    x=r*m.cos(theta)
    y=r*m.sin(theta)
    return x,y

# User input cartesian  coordinates
coordinates=input("Enter the Coordinates: ")
coordinates_list=[int(element) for element in coordinates.split()]

# Function call and storing value of result1 
result1=cartesian_to_polar(*coordinates_list)

# Function call and storing value of result2
result2=polar_to_cartesian(*result1)

# Display result1 and result2
print(f'Polar coordiates radian and theta are :{ result1}')
print(f'Cartesian coordinates x and y are :{result2}')
```

##### The distance between the two points.

```python
import math as m

# Function calculates distance between two points using polar coordinates
def distance_b_two_points(polar_c1,polar_c2):
    
    r1,theta1=polar_c1 # polar coordinates
    r2,theta2=polar_c2

    # Calculate distance 
    distance=round(m.sqrt(r1**2+r2**2-2*r1*r2*m.cos(theta2-theta1)),2)
    return distance
    
# User input polar  coordinates of point1
p_coordinates1=input('Enter the radian and theta:')
p_coordinates_list1=[float(coordinates) for coordinates in p_coordinates1.split()]

# User input polar  coordinates point2
p_coordinates2=input('Enter the radian and theta:')
p_coordinates_list2=[float(coordinates) for coordinates in p_coordinates2.split()]

# Function call
result=distance_b_two_points(p_coordinates_list1,p_coordinates_list2)
print(f'Distance between two points is : {result}')
```

##### Whether the vectors coincide or not. Whether the vectors are orthogonal or not.

```python
def coincide_or_not(polar_c1,polar_c2):
    
    r1,theta1=polar_c1 # polar coordinates
    r2,theta2=polar_c2
    
    # Check are radians and angles same
    if(r1==r2 and theta1==theta2):
        return 'coincide'
        
    return 'do not coincide'
    
def orthogonal_or_not(polar_c1,polar_c2):
    
    r1,theta1=polar_c1 # polar coordinates
    r2,theta2=polar_c2
    angle=theta2-theta1
    
    if(angle<0):
        angle=angle*(-1)
        if(angle==90):
            return 'orthogonal'
        else:
            return 'not orthogonal'
    elif(angle==90):
        return 'orthogonal'
        
    return 'not orthogonal'
    
# User input polar coordinates of vector1
p_coordinates1=input('Enter the radian and theta:')
p_coordinates_list1=[float(coordinates) for coordinates in p_coordinates1.split()]

# User input polar coordinates vector2
p_coordinates2=input('Enter the radian and theta:')
p_coordinates_list2=[float(coordinates) for coordinates in p_coordinates2.split()]

# Function call
result1=coincide_or_not(p_coordinates_list1,p_coordinates_list2)
result2=orthogonal_or_not(p_coordinates_list1,p_coordinates_list2)
print(f'Two vectors {result1} with each other ')
print(f'Two vectors are {result2} to each other ')
```

#### b.2 Going to 3D spherical coordinates
##### Given a point in 3D in Cartesian coordinate convert it to spherical coordinates and viceversa

```python
import numpy as np

# Function calculates spherical coordinates from cartesian coordinates
def cartesian_to_spherical(*c):
    
    x,y,z=c # cartesain coordinates

    # Calculates radian
    r=round(np.sqrt(x**2+y**2+z**2),2)
    
    # Calculates theta and psi
    theta=round((np.arccos(x/r))*(180/np.pi),2)
    psi=round(np.arccos(x/(r*np.sin(theta)))*(180/np.pi),2)
   
    return r,theta,psi

#  Function calculates cartesian coordinates from spherical coordinates
def spherical_to_cartesian(*r):
    r,theta,psi=r[0],r[1],r[2] # spherical coordinates
   
    # Calculates x and y coordinate
    x=round(r*np.cos(psi)*np.sin(theta),2)
    y=round(r*np.sin(psi)*np.sin(theta),2)
    z=round(r*np.cos(theta),2)
    return x,y,z

# sample input dimension
dimension=3

# User input cartesian  coordinates
coordinates=input("Enter the Coordinates: ")
coordinates_list=[int(element) for element in coordinates.split()]

# Function call and storing value of result1 
result1=cartesian_to_spherical(*coordinates_list)

# Function call and storing value of result2
result2=spherical_to_cartesian(*result1)

# Display result1 and result2
print(f'Spherical coordiates radian, theta and psi are :{ result1}')
print(f'Cartesian coordinates x, y and z are :{result2}')
```

##### Given a point in 3D in Cartesian coordinate convert it to cylindrical coordinates and viceversa

```python
import math as m

# Function calculates cylindrical coordinates from cartesian coordinates
def cartesian_to_cylindrical(*c):
    
    x,y,z=c # cartesain coordinates

    # Calculates radian
    r=round(m.sqrt(x**2+y**2),2)
    
    # Calculates theta
    theta=round((m.atan(y/x))*(180/np.pi),2)
   
    return r,theta,z

#  Function calculates cartesian coordinates from cylindrical coordinates
def cylindrical_to_cartesian(*r):
    
    r,theta,z=r # cylindrical coordinates
   
    # Calculates x and y coordinate
    x=round(r*m.cos(theta),2)
    y=round(r*m.sin(theta),2)
    
    return x,y,z

# User input cartesian  coordinates
coordinates=input("Enter the Coordinates: ")
coordinates_list=[int(element) for element in coordinates.split()]

# Function call and storing value of result1 
result1=cartesian_to_cylindrical(*coordinates_list)

# Function call and storing value of result2
result2=cylindrical_to_cartesian(*result1)

# Display result1 and result2
print(f'cylindrical coordiates radian, theta and z are :{ result1}')
print(f'Cartesian coordinates x, y and z are :{result2}')
```
